```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

fig = plt.figure(figsize = [12,6])
df = pd.read_csv("Plot_daily.csv")
df2 = pd.read_csv("Plot_daily_values.csv")
labels =[x[8:] + "-" + x[5:7] for x in df["date"][::5].values]
sns.lineplot(x="date", y="value", data=df)
sns.lineplot(x="date", y="value", data=df2)
locs, labels = plt.xticks(rotation=45, ticks = df["date"][::5].values, labels = labels)
plt.grid()
plt.title('Timeseries daily data')
plt.ylabel("Number of daily tickets")
plt.legend(["Forecast","Observation"])
plt.show()
fig.savefig('daily.png', format='png', dpi=300)
```

```python
import numpy as np
```

```python
np.mean(np.sqrt(np.square(df["value"]-df2["value"])))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd


fig = plt.figure(figsize = [12,6])
df = pd.read_csv("Plot_hourly.csv")
df2 = pd.read_csv("Plot_hourly_values.csv")
labels =[x[8:10] + "-" + x[5:7] for x in df["date"][::24].values]
sns.lineplot(x="date", y="value", data=df)
sns.lineplot(x="date", y="value", data=df2)
locs, labels = plt.xticks(rotation=45, ticks = df["date"][::24].values, labels = labels)
plt.grid()
plt.title('Timeseries hourly data')
plt.ylabel("Number of tickets per hour")
plt.legend(["Forecast","Observation"])
plt.show()
fig.savefig('hourly.png', format='png', dpi=300)
```
